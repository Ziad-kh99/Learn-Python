```python
# list:

lst = [1, 2, 3, 4]

dt_lst = [1, "One", True, 1.0]

print(lst, lst[2], type(lst))
print(dt_lst, dt_lst[1])


# Slicing list:
mylist = [10, 20, 30, 40, 50, 60, 70, 80, 90]

print(mylist[:3])       # [:3] => 0:2
print(mylist[4:9])      # [4:9] => 4:8
print(mylist[-1])
print(len(mylist))



# Manipulate the list:
lst = ['one', 'two', 'three', 4]

print(lst)
lst[len(lst)-1] = 'four'
print(lst)
```

```python
# Iterate over lists:

for i in lst:
    print(i)
```

```python
# list operations 2:

lst = [10,20,30,40,50]

# append item at the end of a list:
lst.append(60)

# insert item in a specific index in a list:
lst.insert(3, 35)

# remove a specific value in a list (first match):
lst.remove(35)

# remove a value in specific index (and store it in a var):
# note that default index (if not specified is -1 (last element))
val = lst.pop(1)


# get the index of a specific value (first match):
idx = lst.index(60)

# sort a list:
lst = [17,8,12,14,5,10]

lst.sort(reverse=True)                    
print(sorted(lst))
print(lst)
```

```python
# Tuple: it's some kind of sequences liks list 
# but it's immutable (we can't change or manipulate its values once defined)

t = (1, 3, 5, 7, 9)

print(type(t))
print(t[3])

# cating tuple to list (to manipulate it)
lst = list(t)

lst[0] = 10

lst
```

```python
# dictionary (dect):

person = {'name': 'ziad', 'age': 26, 'title': 'IT Coordinator'}

print(type(person))
print(person)
print(person['title'])

# add key-value pair:
person['address'] = 'Egypt - Aswan - Edfu'

# list all keys in dict (as dict_keys object):
keys = person.keys()

print(type(keys))       # dict_keys
print(keys)
keys_lst = list(keys)       # casting dict_keys to list
print(keys_lst)

# list all values in dict (as dict_values):
values = person.values()

print(type(values))
print(values)
values_lst = list(values)       # casting dict_values to list
print(values_lst)


# list dict as a list of tuples(key-value pairs)
items = person.items()

print(type(items))
print(items)
items_lst = list(items)     # casting dict_items to list of tuples
print(items_lst)        
print(type(items_lst))


```
