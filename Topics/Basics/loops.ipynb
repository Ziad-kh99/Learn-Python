```python
# while loop: used with condition-based loops.
x = 0

while x <= 10:
    print(x)
    x += 1
```

```python
# for loop: used to iterate over a sequences (lists).

for x in range(5):      # range(5) => [0, 1, 2, 3, 4]
    print(x)

for x in range(1, 7):   # range(1, 7) => [1, 2, 3, 4, 5, 6]
    print(x)
```

```python
# loop statement with 'break':

x = 10

while x > 0:
    if x == 4:
        break       # 'break' will break the loop.
    print(x)
    x -= 1
```

```python
# loop statement with 'continue':

for x in range(1, 11):
    if x < 6:
        continue    # 'continue' will skip the rest of the current iteration.
    print(x)
```
