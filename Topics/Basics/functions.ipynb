```python
# Function: a block of 'resuable' code.

def greeting():
    print('Hello there,')


greeting()

def add(num1, num2):
    print(num1 + num2)

add(1, 5)

# variable number of arguments (tuple)
def add_range(*numbers):
    print(type(numbers))    # *numbers is a 'tuple'
    result = 0
    for i in numbers:
        result += int(i)
    return result

x = add_range(1, 2, 5, 7, '2', 1.5)
print(x)
```
