```python
# naive code:


num1 = 5
num2 = 0
num3 = num1 / num2              # this will raise an ZeroDivisionError.

print('End of the program')     # this statement won't be reached.
```

```python
# try - except

num1 = 5
num2 = 0

try:
    num3 = num1 / num2
except:
    print('Invalid mathimatical operation.')

print('End of the program.')
```

```python
# try - except - else:

num1 = 5
num2 = 0
try:
    num3 = num1 / num2
except:
    print('Invalid mathimatical operation')
else:
    print(num3)


print('End of the program')
```
