```python
# Define a variable:

x = 10
y = 'Hello'
z = False
```

```python
# Python is Dynamiclly-Typed Language:
# We don't need to define static type for the variable and we can change it anytime.

x = 15
x = 20
x = "Welcome"
x = True
```

```python
# String (str):

# var1 = "Hello Python"
var2 = 'Hello Python..'
var3 = '''
This is
multi-line
string
'''


print(var3)
```

```python
# Integer (int):

var1 = 500

print(var1)
```

```python
# Floating Point Numbers (float):

var1 = 1.6
var2 = 12.3456789

print(type(var2))
```

```python
# Boolian (bool):

var1 = True
var2 = False


print(var1, var2, type(var2))
```

```python
# Casting:

x = "99"

# str to int => str(1234)
print(x + str(11))

# int to str => int("1234")
print(int(x) / 3) 

# bool to int:
var3 = True
var4 = False
print(int(var3))
print(int(var4))

# int to bool => bool(1) 
print(bool(0))
```
