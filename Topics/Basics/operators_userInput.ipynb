```python
# Arithmatic Operators ( + - * / // % ** ):

print(5 + 6)
print(7 - 13)
print(5* 7)
print(12 / 4)
print(5 // 2)
print(14 % 4)
print(2 ** 3)


```

```python
# Assignment Operator ( = ):

x = 7
y = x + 3

x += 3      #> x = x + 3
x *= 10     #> x = x * 10

print(x)
```

```python
# Comparison Operators ( < > <= >= == != ): => bool

print(7 == 7)
print(4 != 3)
print(12 >= 15)
```

```python
# Logical Operators ( and or not ):

print(True or False)
print(True and True)
print(not True)

print(False and False or True and True)
```
